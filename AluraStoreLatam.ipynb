### 1. Análisis de facturación



```python
ingresos_lista = [tienda['Precio'].sum() for tienda in tiendas]
nombres_tiendas = [f'Tienda {i+1}' for i in range(len(tiendas))]
ingresos_por_tienda = pd.DataFrame(data=[ingresos_lista], columns=nombres_tiendas, index=['Ingresos'])
ingresos_por_tienda.T.style.format('{:,.0f}')
```

### 2. Ventas por categoría

```python
venta_categoria_por_tienda = [tienda['Categoria'].value_counts() for tienda in tiendas]
venta_categoria_por_tienda = pd.concat(venta_categoria_por_tienda, axis=1, keys=nombres_tiendas)
venta_categoria_por_tienda
```

### 3. Calificación promedio de la tienda


```python
cal_promedio_tienda = [tienda['Calificacion'].mean() for tienda in tiendas]
cal_promedio_tienda = pd.DataFrame(data=[cal_promedio_tienda], columns=nombres_tiendas, index=['Promedio'])
cal_promedio_tienda.T
```

### 4. Productos más y menos vendidos

```python
venta_categoria_por_tienda
```

```python
lista_prod_mas_vendido = [tienda['Categoria'].value_counts().idxmax() for tienda in tiendas]
lista_prod_menos_vendido = [tienda['Categoria'].value_counts().idxmin() for tienda in tiendas]
prod_mas_y_menos_vendidos = pd.DataFrame(data=[lista_prod_mas_vendido, lista_prod_menos_vendido], columns=nombres_tiendas, index=['Prod mas vendidos', 'Prod menos vendidos'])
prod_mas_y_menos_vendidos.T
```

### 5. Envío promedio por tienda

```python
costo_envio_prom = [tienda['Costo_envio'].mean() for tienda in tiendas]
costo_envio_prom = pd.DataFrame(data=[costo_envio_prom], columns=nombres_tiendas, index=['Media_Costo_Envio'])
costo_envio_prom.T
```

## **Visualizacion**

### **Herramientas para la Creacion de las Graficas**

#### Librerias

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

#### Formateadores de Cantidades (millones y miles)

```python
formatter_millones = ticker.FuncFormatter(lambda x, pos: f'{x/1e6:.0f}M')
formatter_miles = ticker.FuncFormatter(lambda x, pos: f'{x/1e3:.0f}K')
```

#### Extractor de "Series" de cada DataFrame

```python
datos_ingresos = ingresos_por_tienda.loc['Ingresos']
datos_costos = costo_envio_prom.loc['Media_Costo_Envio']
datos_calificacion = cal_promedio_tienda.loc['Promedio']
```

#### Costo de Envio y Ventas por Tienda

```python
ax = datos_costos.plot(kind='bar', figsize=(12, 8), color='red', ylabel='Costo Promedio de Envio')
ax.set_title("Costo de Envio Promedio por Tienda")
ax.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

##### Grafica lineal - Ventas en Millones

```python
ax = datos_ingresos.plot(kind='line', figsize=(12,8), ylabel='Ventas (en Millones)', marker='o', lw=2, color='blue')
ax.yaxis.set_major_formatter(formatter_millones)
ax.set_title('Ventas Totales por Tienda', fontsize=16)
ax.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

#### Ganancias Totales vs. Costo Promedio de Envio

```python
df_comparativo = pd.DataFrame({
    'Ganancias': datos_ingresos,
    'Costo de Envío': datos_costos
})

fig, ax = plt.subplots(figsize=(15, 8))

color_bar = 'royalblue'
df_comparativo['Ganancias'].plot(kind='bar', ax=ax, color=color_bar, label='Ganancias')
ax.set_ylabel('Ganancias (en Millones)', color=color_bar)
ax.tick_params(axis='y', labelcolor=color_bar)
ax.yaxis.set_major_formatter(formatter_millones)
ax.tick_params(axis='x', rotation=0)

max_ganancias = df_comparativo['Ganancias'].max()
ax.set_ylim(bottom=0, top=max_ganancias * 1.15)

ax2 = ax.twinx()
color_line = 'black'
df_comparativo['Costo de Envío'].plot(kind='line', ax=ax2, color=color_line, marker='o', label='Costo de Envío (Línea)', zorder=2)
ax2.set_ylabel('Costo de Envío (en Miles)', color=color_line)

max_costo_envio = df_comparativo['Costo de Envío'].max()
ax2.set_ylim(bottom=12000, top=ax.get_ylim()[1] * (max_costo_envio / max_ganancias))

ax2.yaxis.set_major_formatter(formatter_miles)

ax.set_title('Ganancias Totales (Barras) vs. Costo Promedio de Envío (Linea)', fontsize=16)
ax.grid(True, linestyle='--', alpha=0.8)
#ax2.grid(True, linestyle=':', alpha=0.4)

fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.95))

plt.tight_layout()
plt.show()
```

## **Informe Final**

### **Informe de Análisis y Recomendación de Venta de Tiendas**

El presente informe detalla un analisis exhaustivo de las cuatro tiendas de su propiedad. El objetivo es identificar que tienda representa la mejor candidata para una venta, basando la decisión en un equilibrio entre el rendimiento financiero y la satisfaccion del cliente.

1. **Introduccion**


Para ello, se analizaron cinco factores clave, respaldados por los datos y graficos generados:
*   Los ingresos totales de las tiendas.
*   Las categorias de productos mas y menos vendidas.
*   Las calificaciones promedio de los clientes por tienda.
*   Los productos mas y menos vendidos.
*   El coste de envio promedio para cada tienda.

2. **Desarrollo y Analisis de Factores**


Podemos empezar a destacar algunos puntos clave para determinar que tienda es mas conveniente realizar su venta (cabe aclarar que para este analisis estamos tomando en cuenta todo lo registrado, sin importar las fechas). Quiero destacar que en general, la venta de productos por categoria considero que es bastante balanceada y que no veo que algun producto en especifico (hablando de la comparacion por tienda) destaque por sobre otros productos, con esto me refiero a que si en algun momento se tiene pensado elegir una tienda por que X producto destaca mas o menos, no hay argumento claro para que esta sea la razon, por lo tanto, descartare este analisis a futuro (esta informacion se puede ver mas claramente en la Grafica "Distribucion de Ventas por Tienda").
Ahora me gustaria ir punto por punto, hablando de los otros factores restantes para discutir en que se destaca o falla cada tienda.

**A. Ingresos Totales por Tienda (Grafico: "Distribucion de Ventas")**

El primer factor analizado fueron los ingresos brutos. La "Tienda 1" es la lider indiscutible en generacion de ingresos, abarcando el 26.1% del total. Por el contrario, la "Tienda 4" es la que menos ingresos genera, con un 23.6%.

* Tienda 1: 26.1% (Lider en ingresos)
* Tienda 2: 25.4% (Solido)
* Tienda 3: 24.9% (Solido)
* Tienda 4: 23.6% (Mas baja en ingresos)

Este punto posiciona a la Tienda 1 como el activo mas valioso en términos de facturacion pura.

**B. Analisis por Categoria (Grafico: "Ventas por Categoria")**

Si bien las ventas pueden parecer balanceadas, un analisis mas detallado del grafico "Ventas por Categoria en cada Tienda" revela matices importantes:

* Categorias Fuertes: "Muebles" y "Electronicos" son, con diferencia, las categorias que mas ventas generan en todas las tiendas.

* Lideres de Nicho: La "Tienda 3" muestra una fortaleza particular en "Muebles". La "Tienda 4", aunque mas baja en ingresos totales, destaca en "Articulos para el hogar".

* Conclusion de Categoria: No hay una tienda que domine o falle en todas las categorias. Cada una tiene sus fortalezas y debilidades, lo que sugiere que el portafolio de productos esta razonablemente diversificado en las cuatro ubicaciones.

**C. Calificaciones Promedio (Grafico: "Calificacion Promedio")**

La satisfaccion del cliente, medida en calificaciones (1-5 estrellas), muestra una historia opuesta a la de los ingresos.

* Mejores Calificadas: "Tienda 3" y "Tienda 4" lideran la satisfaccion del cliente, siendo ligeramente superior la "Tienda 3", superando ambas la marca de 4 estrellas.
* Peores Calificada: La "Tienda 1", a pesar de ser la que mas vende, recibe la calificacion promedio mas baja, y la "Tienda 2" no se queda muy lejos, ambas sin siquiera llegar a las 4 estrellas.

Esto demuestra una clara desconexion entre volumen de ventas y satisfaccion del cliente en la Tienda 1.

**D. Costo de Envio Promedio (Grafico: "Ganancias vs. Costo Promedio")**

Este factor es crucial para la experiencia del cliente. El analisis (visible en la linea negra del grafico combinado) revela:

Mejor para el Cliente: La "Tienda 4" ofrece el costo de envio promedio mas bajo (aprox. 23,500), lo cual es un gran atractivo y puede explicar en parte sus buenas calificaciones.

Peor para el Cliente: La "Tienda 1" tiene el costo de envio promedio mas alto (aprox. 26,000). Este costo elevado es el punto que, con seguridad, impacta negativamente en sus calificaciones.

Punto Medio: Las tiendas 2 y 3 se mantienen en un rango intermedio aceptable (aprox. 24,800 - 25,200).

3. **Conclusiones y Recomendacion Final**


La decision de que tienda vender se reduce a dos enfoques estrategicos distintos, ya que el analisis revela dos tiendas (Tienda 1 y Tienda 4) que son opuestas en rendimiento.

* Enfoque 1: Maximizar la Ganancia de la Venta (Vender la Tienda 1)
Si el objetivo es obtener el maximo capital de la venta, la "Tienda 1" es la candidata. Es la que mas factura (26.1%) y podria tener la valoracion mas alta. Sin embargo, esto significa conservar las tiendas con menores ingresos.

* Enfoque 2: Optimizar el Portafolio y la Satisfaccion (Vender la Tienda 4)
Si el objetivo es "recortar" el activo de menor rendimiento financiero, la "Tienda 4" seria la eleccion, ya que genera los ingresos mas bajos (23.6%).

* Enfoque 3: Optimizar la Calidad del Negocio (Vender la Tienda 1)
Este enfoque, que prioriza la sostenibilidad a largo plazo y la reputacion de la marca, ve las cosas de manera diferente. La "Tienda 1" es la mas problematica:

  1. Tiene la peor calificacion de clientes.
  2. Tiene el costo de envio mas alto (lo que genera friccion).

Es un activo que genera altos ingresos, pero a costa de una alta insatisfaccion. Esto la convierte en un activo complicado y que puede ser mas dificil de manejar (en reputacion y logistica). Las tiendas 2 y 3 son las mas balanceadas (buenos ingresos, buenas calificaciones) y la Tienda 4, aunque con menos ingresos, tiene clientes mas felices y costos de envio mas bajos, lo que sugiere un potencial de crecimiento mas saludable.

**Recomendacion Profesional**

Desde un punto de vista estrategico que prioriza la salud de la marca y la satisfaccion del cliente (que a largo plazo impulsa el crecimiento sostenible), la tienda mas prescindible es aquella que daña la reputacion del grupo.

**Recomendacion: Vender la Tienda 1.**

Justificacion: A pesar de sus altos ingresos, la "Tienda 1" es la que peor rendimiento tiene en metricas de satisfaccion (calificaciones mas bajas) y eficiencia operativa (costos de envio mas altos). Venderla permite al Sr. Juan concentrar sus esfuerzos en las tiendas 2 y 3 (las mas balanceadas y eficientes) y en la Tienda 4 (con alto potencial de crecimiento gracias a su alta satisfaccion de cliente y bajos costos), ademas, al ser la que mas produce, otorgara mayores ganancias por su venta, dando la posibilidad de reinvertir y mejorar el resto de tiendas.
